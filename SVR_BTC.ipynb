```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("bitcoin.xlsx")
df.head(10)
```

```python
df.describe()
```

```python
X = df.copy()
X = X.drop(columns=['BTC', 'Date'])

y = df['BTC'].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
mean_X_train=np.mean(X_train)
std_X_train=np.std(X_train)
# print(std_X_train)
X_train = (X_train - mean_X_train)/std_X_train


mean_y_train=np.mean(y_train)
std_y_train=np.std(y_train)
y_train=(y_train-mean_y_train)/std_y_train
y_train


## apply same transformation to test data
X_test = (X_test-mean_X_train)/std_X_train  # remove mean and standard deviation from test
y_test = (y_test-mean_y_train)/std_y_train
X_train
```

```python
!pip install python-math
import math
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
C_range = np.arange(1, 101, 5)
epsilon_range = np.arange(0.01,0.1,0.01)


param_grid = dict(C = C_range, epsilon = epsilon_range, )
grid = GridSearchCV(SVR(kernel='linear'), param_grid = param_grid, cv = 3,  n_jobs = 7)  #  3-fold cross-validation
model = grid.fit(X_train, y_train)
print("The best parameters are %s with a score of %0.2f" % (grid.best_params_, grid.best_score_))

clf = SVR(kernel='linear', C = grid.best_params_["C"], epsilon = grid.best_params_["epsilon"])
model_best = clf.fit(X_train, y_train)

y_hat_in_svr_linear = model_best.predict(X_train)
y_hat_out_svr_linear = model_best.predict(X_test)

print("Residual sum of squares: %.3f"
      % np.mean((y_hat_in_svr_linear - y_train) ** 2))
print("Root Mean Square (in-sample): %.3f"
      % math.sqrt(np.mean((y_hat_in_svr_linear - y_train)**2)))

print("Residual sum of squares: %.3f"
      % np.mean((y_hat_out_svr_linear - y_test) ** 2))
print("Root Mean Square (in-sample): %.3f"
      % math.sqrt(np.mean((y_hat_out_svr_linear - y_test)**2)))
```

```python
C_range = np.arange(1, 101, 5)
gamma_range=np.arange(1, 101, 5)
epsilon_range = np.arange(0.01,0.1,0.01)

param_grid = dict(gamma=gamma_range, C=C_range, epsilon=epsilon_range, )
grid = GridSearchCV(SVR(kernel='rbf'), param_grid=param_grid, cv=5,  n_jobs=7)  #  5-fold cross-validation
model = grid.fit(X_train, y_train)
print("The best parameters are %s with a score of %0.2f" % (grid.best_params_, grid.best_score_))

clf2=SVR(kernel='rbf', C=grid.best_params_["C"], epsilon=grid.best_params_["epsilon"], gamma=grid.best_params_["gamma"])
model_best=clf2.fit(X_train, y_train)

y_hat_in_svr_rbf = model_best.predict(X_train)
y_hat_out_svr_rbf = model_best.predict(X_test)

print("Residual sum of squares: %.3f"
      % np.mean((y_hat_in_svr_rbf - y_train) ** 2))
print("Root Mean Square (in-sample): %.3f"
      % math.sqrt(np.mean((y_hat_in_svr_rbf - y_train)**2)))

print("Residual sum of squares: %.3f"
      % np.mean((y_hat_out_svr_rbf - y_test) ** 2))
print("Root Mean Square (out-of-sample): %.3f"
      % math.sqrt(np.mean((y_hat_out_svr_rbf - y_test)**2)))
```

```python
print('------------------------SVR-Linear---------------------')
print("Root Mean Square (in-sample):  %.3f"
      % math.sqrt(np.mean((y_hat_in_svr_linear - y_train)**2)))
print("Root Mean Square (out-of-sample):  %.3f"
      % math.sqrt(np.mean((y_hat_out_svr_linear - y_test)**2)))
print('------------------------SVR-RBF------------------------')
print("Root Mean Square (in-sample):  %.3f"
      % math.sqrt(np.mean((y_hat_in_svr_rbf - y_train)**2)))
print("Root Mean Square (out-of-sample):  %.3f"
      % math.sqrt(np.mean((y_hat_out_svr_rbf - y_test)**2)))
print('-------------------------------------------------------')
```

```python
import itertools
X_new = df[['eur/usd','gbp/usb','yen/usd']].copy()
y_new = df['BTC'].copy()

X_new = X_new.fillna(X_new.mean())   #replace NaNs with the mean value
y_new = y_new.fillna(y_new.mean())   #replace NaNs with the mean value


X_train_new, X_test_new, y_train_new, y_test_new = train_test_split( X_new, y_new, test_size=0.2, random_state=4)

mean_X_train_new=np.mean(X_train_new)
std_X_train_new=np.std(X_train_new)
X_train_new=(X_train_new-mean_X_train_new)/std_X_train_new

mean_y_train_new=np.mean(y_train_new)
std_y_train_new=np.std(y_train_new)
y_train_new=(y_train_new-mean_y_train_new)/std_y_train_new

## apply same transformation to test data
X_test_new = (X_test_new-mean_X_train_new)/std_X_train_new  # remove mean and standard deviation from test
y_test_new = (y_test_new-mean_y_train_new)/std_y_train_new


C_range = np.arange(1, 11, 5)
gamma_range=np.arange(1, 11, 5)
epsilon_range = np.arange(0.01,0.02,0.1)

param_grid = dict(gamma=gamma_range, C=C_range, epsilon=epsilon_range, )
grid = GridSearchCV(SVR(kernel='rbf'), param_grid=param_grid, cv=5,  n_jobs=7)  #  5-fold cross-validation

featureList =  ['eur/usd','gbp/usb','yen/usd']
numInputs = 2
AllParms = []
AllParms = list(itertools.combinations(featureList, numInputs))

# Create a list of integers to identify the iteration, i1, i2, i3, etc.
iteration = list(range(1, len(AllParms) + 1))

rmse=[]

# Loop thru files
for j in range(len(AllParms)):
    nAllParms=list(AllParms[j]) 
    model = grid.fit(X_train_new[nAllParms], y_train_new)
    y_out = model.predict(X_test_new[nAllParms])
    rmse.append(math.sqrt(np.mean((y_out - y_test_new)**2)))
    
print(rmse)    
```
